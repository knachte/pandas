```python
fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(df["MONTH"], df["DATA1"])
ax[0, 1].plot(df["MONTH"], df["DATA2"])
ax[1, 0].plot(df["MONTH"], df["DATA3"])
ax[1, 1].plot(df["MONTH"], df["DATA4"])

plt.show()
```

```python
# Create a figure and an array of axes: 2 rows, 1 column with shared y axis
fig, axes = plt.subplots(2, 1, sharey=True)

# Plot Seattle precipitation data in the top axes
axes[0].plot(df["MONTH"], df["DATA1"], color = "b")
axes[0].plot(df["MONTH"], df["DATA2"], color = "b", linestyle = "--")
axes[0].tick_params('y', colors="b")

# Plot Austin precipitation data in the bottom axes
axes[1].plot(df["MONTH"], df["DATA3"], color = "r")
axes[1].plot(df["MONTH"], df["DATA4"], color = "r", linestyle = "--")
axes[1].tick_params('y', colors="r")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(df["MONTH"], df["DATA1"], color = "r")
ax.set_ylabel("red line", color="r")

# Create a twin Axes object that shares the x-axis
ax2 = ax.twinx()

ax2.plot(df["MONTH"], df["DATA5"], color = "b")
ax2.set_ylabel("blue line", color="b")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(df["MONTH"], df["DATA1"], color = "r")
ax.set_ylabel("red line", color="r")
ax.annotate("halfway", xy=(6, 7), xytext=(5, 3), arrowprops={"arrowstyle":"->", "color":"gray"})
```

```python
fig, ax = plt.subplots()

ax.bar(df.MONTH, df.DATA1, label="Data1")
ax.bar(df.MONTH, df.DATA2, bottom=df.DATA1, label="Data2")
ax.bar(df.MONTH, df.DATA3, bottom=df.DATA1+df.DATA2, label="Data3")

# Display the legend
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(df.DATA3, label="Data3", histtype="step", bins=5)
ax.hist(df.DATA2, label="Data2", histtype="step", bins=5)

# Add the legend and show the Figure
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar("Rowing", df.DATA2.mean(), yerr=df.DATA2.std())
ax.bar("Gymnastics", df.DATA3.mean(), yerr=df.DATA3.std())

plt.show()
```

```python
fig, ax = plt.subplots()

# Add Seattle temperature data in each month with error bars
ax.errorbar(df["MONTH"], df["DATA2"], yerr = df["DATA6"])

# Add Austin temperature data in each month with error bars
ax.errorbar(df["MONTH"], df["DATA3"], yerr = df["DATA6"]) 

# Set the y-axis label
ax.set_ylabel("Temperature (Fahrenheit)")

plt.show()
```

```python
fig, ax = plt.subplots()

# Add a boxplot for the "Height" column in the DataFrames
ax.boxplot([df["DATA2"],df["DATA3"]])

# Add x-axis tick labels:
ax.set_xticklabels(["DATA2", "DATA3"])

# Add a y-axis label
ax.set_ylabel("Label")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(df.DATA2, df.DATA3, c=df.DATA1)

plt.show()
```
